## Matplotlib - Scatterplot

![](images/matplot.png)

The provided Python code uses Matplotlib to generate a scatter plot with random data, custom sizes, and colors, while also setting specific limits and ticks for both x and y axes.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8),
       xlabel="x-axis", ylabel="y-axis",
       title="Matplotlib-Scatterplot")
plt.grid(alpha=0.25)
plt.show()
```

## Seaborn - Boxplot

![](images/seaborng.png)

The provided Python code utilizes Seaborn to create a nested boxplot displaying the distribution of total bills based on day, time, and smoker status using the "tips" dataset, with a customized color palette and additional styling features.

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

sns.set_theme(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
sns.despine(offset=10, trim=True)
plt.title("Bills by Day vs Time")
plt.grid(alpha=0.25)
plt.show()
```

## Plotly - Barplot

![](images/Plotly-logo.png)

The provided Python code utilizes Plotly Express to generate an interactive bar chart using long-form data, specifically displaying medal counts by nation and medal type. The visualization is designed for compatibility with notebook environments using Plotly's offline mode.

```python
import plotly.express as px
from plotly.offline import init_notebook_mode

init_notebook_mode(connected=True)   
  
long_df = px.data.medals_long()

fig = px.bar(long_df, x="nation", y="count", color="medal", title="Long-Form Input")
fig.show()
```
